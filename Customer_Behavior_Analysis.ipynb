# to replace null i am using median because for outliers it will give good result

# if we use median for review lets say clothing then it will not a good choice because data will look like synthetic

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# checking null values
print(df.isnull().sum())
```

```python
# instead of writing column name as camel case and have smpacce between them we can write column name in snake case and have underscore between them

# method1 to replace space with underscore
# df.columns=[col.lower().replace(' ','_')for col in df.columns]
# print(df.columns)

# method2 to replace space with underscore
df.columns=df.columns.str.lower().str.replace(' ','_')
print(df.columns)
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# create a column age group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
print(df[['age','age_group']].head(10))
```

```python
# create column for frequency days
frequency_mapping={'Fortnightly':14,'Weekly':7,'Monthly':30, 'Quarterly':90,'Annually':365,'Bi-Weekly':14,'Every 3 Months':90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
print(df[['frequency_of_purchases','purchase_frequency_days']].head(10))
```

```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
import re

# Connection details
host = "localhost"
database = "customer_behavior"
user = "postgres"
password = "postgres123"
port = "5432"

# Create engine
engine = create_engine(
    f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}"
)

# Load data
df = pd.read_csv("customer_shopping_behavior.csv")

# 🔥 Clean column names (camelCase + spaces → snake_case)
def clean_column(col):
    col = col.strip()
    col = re.sub(r'([a-z])([A-Z])', r'\1_\2', col)  # camelCase → snake_case
    col = col.replace(" ", "_")
    return col.lower()

df.columns = [clean_column(col) for col in df.columns]

# Insert into PostgreSQL
table_name = 'customer_table'
df.to_sql(table_name, engine, if_exists='replace', index=False)

print(f"Data inserted into {table_name} table successfully.")

print(df.head())
```
